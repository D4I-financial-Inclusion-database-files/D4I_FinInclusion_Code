## Imports:

```python
import requests
import json
import pandas as pd
import numpy as np
```

## Input the Data:

```python
about_COVID_api = requests.get("https://corona-api.com/countries")
about_COVID_api
```

Response 200 means it is an active API

After this we cover to json, thats how python can better read APIs

```python
covid_dict = about_COVID_api.json()
covid_dict
```

```python
covid_dict.keys()
```

```python
covid_dict['data']
```

```python
df = pd.DataFrame(covid_dict['data'])
df
```

## Data Manipulation:

```python
#lat/long
lat_long = df['coordinates'].apply(pd.Series)
#deaths today
today_info = df['today'].apply(pd.Series)
#latest data
latest_data = df['latest_data'].apply(pd.Series)
latest_data
```

```python
latest_expanded = pd.concat([latest_data.drop(['calculated'], axis=1), latest_data['calculated'].apply(pd.Series)], axis=1)
latest_expanded
```

```python
df2 = df.drop(['today','coordinates',"latest_data"], axis=1)

df3 = pd.concat( [df2,lat_long,today_info,latest_expanded], axis=1)
df3
```
